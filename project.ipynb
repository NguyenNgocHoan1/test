**1.3 Cơ sở lý thuyết về phân cụm (Clustering)**

**1.3.1. Tổng quan về bài toán phân cụm**

- Khái niệm: Phân cụm là kỹ thuật học máy không giám sát (Unsupervised Learning) nhằm nhóm các đối tượng dữ liệu có đặc điểm tương đồng vào cùng một nhóm (cluster), sao cho sự tương đồng trong cùng một nhóm là lớn nhất và giữa các nhóm là nhỏ nhất.

- Ứng dụng: Trong du lịch, phân cụm giúp:

    - Phân khúc thị trường (Market Segmentation).

    - Gợi ý gói du lịch phù hợp (Recommendation Systems).

    - Tối ưu hóa chiến dịch marketing (Targeted Marketing).

**1.3.2. Giới thiệu thuật toán K-Means**

K-Means là một trong những thuật toán phân cụm phổ biến nhất nhờ sự đơn giản và hiệu quả tính toán trên tập dữ liệu lớn.

**a. Nguyên lý hoạt động:** Thuật toán hướng tới việc tối thiểu hóa hàm mục tiêu là tổng bình phương khoảng cách giữa các điểm dữ liệu và tâm cụm của chúng. Quy trình bao gồm 4 bước lặp:

1. Chọn ngẫu nhiên $K$ điểm làm tâm cụm ban đầu (centroids). ( Trong đồ án, kỹ thuật K-Means++ được sử dụng để tối ưu hóa bước chọn tâm ban đầu nhằm tăng tốc độ hội tụ).

2. Gán mỗi điểm dữ liệu vào tâm cụm gần nhất (dựa trên khoảng cách Euclid).

3. Cập nhật lại vị trí tâm cụm bằng cách tính trung bình tọa độ các điểm trong cụm.

4. Lặp lại bước 2 và 3 cho đến khi tâm cụm không thay đổi hoặc đạt số vòng lặp tối đa.

**b. Hàm mục tiêu (Objective Function):**
$$J = \sum_{j=1}^{K} \sum_{i=1}^{n} ||x_i^{(j)} - c_j||^2$$

Trong đó:
- $K$: Số lượng cụm.
- $x_i^{(j)}$: Điểm dữ liệu thứ $i$ thuộc cụm $j$
- $c_j$: Tâm của cụm $j$.
- $||...||^2$: Bình phương khoảng cách Euclidean

**c. Ưu và nhược điểm**

- Ưu điểm: Đơn giản, tốc độ xử lý nhanh với dữ liệu lớn, dễ giải thích kết quả.

- Hạn chế: Phải xác định trước số cụm K, nhạy cảm với nhiễu (outliers) và dữ liệu không có dạng hình cầu.

**1.3.3. Các phương pháp đánh giá chất lượng cụm**

Vì phân cụm là bài toán học không giám sát (Unsupervised Learning) nên không có "nhãn đúng" để so sánh. Do đó, nghiên cứu sử dụng kết hợp hai phương pháp đánh giá nội tại phổ biến nhất để xác định số lượng cụm tối ưu ($K$):

**a. Phương pháp Elbow (The Elbow Method)** Phương pháp này dựa trên việc tính toán tổng bình phương khoảng cách từ mỗi điểm đến tâm cụm của nó (Within-Cluster Sum of Squares - WCSS)

- Nguyên lý: Khi số lượng cụm $K$ tăng lên, WCSS sẽ giảm xuống vì các cụm trở nên nhỏ hơn và chặt chẽ hơn.

- Tiêu chí chọn K: Chọn giá trị $K$ tại điểm mà tốc độ giảm của WCSS bắt đầu chững lại (tạo thành hình khuỷu tay). Tại điểm này, việc tăng thêm cụm không mang lại hiệu quả nén dữ liệu đáng kể.

- Công thức WCSS:$$WCSS = \sum_{i=1}^{K} \sum_{x \in C_i} ||x - \mu_i||^2$$(Trong đó: $C_i$ là cụm thứ $i$, $x$ là điểm dữ liệu, $\mu_i$ là tâm của cụm $i$).

**b. Chỉ số Silhouette (Silhouette Score)** Chỉ số này đo lường chất lượng của việc phân cụm dựa trên hai yếu tố: độ nén (cohesion) bên trong cụm và độ tách biệt (separation) giữa các cụm.

- Nguyên lý: Với mỗi điểm $i$, hệ số Silhouette $s(i)$ được tính dựa trên khoảng cách trung bình đến các điểm cùng cụm ($a$) và khoảng cách trung bình đến các điểm thuộc cụm lân cận gần nhất ($b$):$$s(i) = \frac{b(i) - a(i)}{\max\{a(i), b(i)\}}$$

- Tiêu chí chọn K: Giá trị trung bình của Silhouette Score trên toàn bộ dữ liệu nằm trong khoảng $[-1, 1]$.
    - Gần 1: Điểm dữ liệu được phân cụm tốt (gần tâm mình, xa tâm khác).

    - Gần 0: Điểm dữ liệu nằm ở biên giới giữa hai cụm.

    - Gần -1: Điểm dữ liệu có thể bị phân sai cụm.

    - Kết luận: Chọn $K$ có điểm Silhouette trung bình cao nhất.

Chiến lược lựa chọn: Nghiên cứu sẽ ưu tiên giá trị $K$ được gợi ý bởi cả hai phương pháp trên để đảm bảo các cụm vừa có độ nén tốt, vừa tách biệt rõ ràng.

**2.4. Phương pháp nghiên cứu**

Để giải quyết bài toán phân cụm khách hàng dựa trên dữ liệu hành vi tiêu dùng, nghiên cứu này áp dụng quy trình khai phá dữ liệu tiêu chuẩn, kết hợp với các kỹ thuật đánh giá định lượng để đảm bảo tính chính xác của mô hình.

**2.4.1. Quy trình phân tích**

Quá trình nghiên cứu được thực hiện qua 5 bước chính, đảm bảo dữ liệu thô được chuyển hóa thành các dữ liệu có giá trị (insights). Sơ đồ quy trình tổng quát như sau:



Chi tiết từng bước thực hiện:

1. **Bước 1: Thu thập và Chọn lọc dữ liệu**

    - Sử dụng bộ dữ liệu `tourism_consumer_behavior` bao gồm thông tin nhân khẩu học và hành vi tiêu dùng của khách du lịch.

    - Kiểm tra và Xử lý các giá trị thiếu (missing values), các dòng trùng lặp(duplicate rows) và kiểm tra tính nhất quán của dữ liệu.

    - Tiến hành lựa chọn các đặc trưng (feature selection) có ý nghĩa phân loại cao như: Độ tuổi, Giới tính, Thu nhập, Loại hình du lịch yêu thích, Kênh đặt vé, Nhu cầu quay lại, Mức độ chi tiêu, Thời lượng mỗi chuyến đi..

2. **Bước 2: Phân tích dữ liệu khám phá (EDA)**

    Bước này tập trung vào việc thấu hiểu toàn diện đặc điểm và hành vi khách hàng thông qua các phương pháp trực quan hóa dữ liệu. Quá trình phân tích được thực hiện theo trình tự từ tổng quát đến chi tiết nhằm phục vụ cho bài toán Phân cụm (Clustering) ở bước sau:
    - **Phân tích đơn biến:** Khảo sát phân phối của từng đặc trưng, tách biệt theo hai nhóm: Nhân khẩu học - Sở thích và Hành vi tiêu dùng để phác họa chân dung khách hàng cơ bản.
    - **Phân tích hai biến:** Đánh giá mối quan hệ giữa các cặp biến định lượng và định tính nhằm kiểm tra các giả thuyết về xu hướng chi tiêu và lòng trung thành.
    - **Phân tích đa biến:** Sử dụng ma trận tương quan (Correlation Matrix) để kiểm tra mức độ độc lập tuyến tính giữa các biến, đảm bảo tính hiệu quả khi đưa vào thuật toán phân cụm.

3. **Bước 3: Tiền xử lý dữ liệu (Data Preprocessing)**

    Đây là bước quan trọng nhằm chuẩn hóa dữ liệu đầu vào, đảm bảo độ chính xác khi áp dụng thuật toán dựa trên khoảng cách như K-Means. Quy trình bao gồm:

    - **Mã hóa biến phân loại (Encoding):** Chuyển đổi các biến định tính (dạng chữ) sang dạng số học bằng các kỹ thuật phù hợp với đặc tính của từng biến (sử dụng kết hợp **One-Hot Encoding** cho biến định danh và **Ordinal Encoding** cho biến thứ bậc).

    - **Chuẩn hóa dữ liệu (Scaling):** Đồng bộ hóa thang đo của các biến số học về cùng một miền giá trị (sử dụng **StandardScaler**) để loại bỏ sự chênh lệch về đơn vị đo, ngăn chặn việc các biến có giá trị lớn chi phối kết quả phân cụm.
    
4. **Bước 4: Xây dựng mô hình phân cụm**

    Giai đoạn này tập trung vào việc xây dựng và tối ưu hóa thuật toán phân cụm:

    - **Lựa chọn thuật toán:** Áp dụng thuật toán K-Means Clustering với kỹ thuật khởi tạo tâm cụm `k-means++`. Phương pháp này giúp chọn các điểm tâm ban đầu một cách tối ưu, tăng tốc độ hội tụ và độ chính xác so với việc khởi tạo ngẫu nhiên.

    - **Xác định số cụm tối ưu ($K$):** Thay vì chọn $K$ theo cảm tính, nghiên cứu sử dụng phương pháp kết hợp hai chỉ số định lượng:
        - Phương pháp Elbow (Inertia): Để tìm điểm bão hòa về độ nén của dữ liệu.
        - Hệ số Silhouette: Để đảm bảo các cụm có độ tách biệt rõ ràng nhất.
    - **Huấn luyện mô hình:** Tiến hành phân nhóm dữ liệu dựa trên giá trị $K$ tối ưu đã xác định và gán nhãn cụm cho từng khách hàng.

5. **Bước 5: Phân tích đặc điểm cụm và đề xuất chiến lược**

    - Phân tích đặc điểm của từng cụm (profiling) để dán nhãn và đề xuất chiến lược kinh doanh phù hợp.






**2.4.2 Phương pháp đánh giá mô hình**

Do phân cụm khách hàng là bài toán Học không giám sát (Unsupervised Learning), dữ liệu đầu vào không có nhãn (Label) hay đáp án đúng (Ground Truth) để so sánh trực tiếp như các bài toán phân loại. Vì vậy, nghiên cứu không thể sử dụng các chỉ số thông thường như Accuracy hay F1-Score.

Thay vào đó, để đảm bảo tính khách quan và chính xác khi xác định số lượng cụm tối ưu ($K$), nghiên cứu áp dụng kết hợp hai phương pháp định lượng dựa trên nền tảng lý thuyết đã trình bày tại **Chương 1 (Mục 1.3.3)**. Quy trình cụ thể gồm các bước:
- Chạy thuật toán K-Means lặp lại với số cụm $K$ thay đổi trong khoảng từ 2 đến 10.
- Tại mỗi giá trị $K$, tính toán đồng thời hai chỉ số:
    - WCSS (cho phương pháp Elbow): Để đánh giá độ nén của dữ liệu (cụm có chặt chẽ hay không).
    - Silhouette Score: Để đánh giá độ tách biệt giữa các cụm (các nhóm khách hàng có thực sự khác biệt hay không).
- Lựa chọn giá trị $K$ tối ưu dựa trên điểm "khuỷu tay" của biểu đồ WCSS và giá trị trung bình cao nhất của Silhouette.

```python
# Tạo khung hình lớn (Figure) chứa 4 biểu đồ con (Axes)
# nrows=2, ncols=2: Chia lưới 2 hàng 2 cột
# figsize=(16, 10): Kích thước tổng thể là 16x10 inch
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 8))

# Đặt tiêu đề chung cho cả tấm hình lớn
fig.suptitle('3.2.1. Phân tích Nhân khẩu học và Sở thích', fontsize=18, fontweight='bold')

# --- BIỂU ĐỒ 1: Phân phối Tuổi (Góc trên trái: axes[0,0]) ---
# sns.histplot: Vẽ biểu đồ phân phối
# kde=True: Vẽ thêm đường cong mật độ (Kernel Density Estimate) để thấy rõ hình dáng phân phối
# bins=20: Chia độ tuổi thành 20 khoảng nhỏ
sns.histplot(data=data, x='Age', kde=True, bins=20, color='blue', ax=axes[0, 0])
axes[0, 0].set_title('Phân phối Độ tuổi (Age Distribution)') # Đặt tên biểu đồ con

# --- BIỂU ĐỒ 2: Tỷ lệ Giới tính (Góc trên phải: axes[0,1]) ---
# Bước chuẩn bị: Pie chart cần dữ liệu tổng hợp, không vẽ trực tiếp từ raw data được
gender_counts = data['Gender'].value_counts()

# ax.pie: Vẽ biểu đồ tròn
# autopct='%1.1f%%': Hiển thị số phần trăm với 1 số lẻ
# startangle=140: Xoay biểu đồ đi một chút cho đẹp
# explode: Tách nhẹ các miếng bánh ra (ở đây tạo khoảng cách 0.05 cho tất cả)
axes[0, 1].pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', 
               startangle=140, colors=sns.color_palette('pastel'), explode=[0.05]*len(gender_counts))
axes[0, 1].set_title('Tỷ lệ Giới tính (Gender Ratio)')

# --- BIỂU ĐỒ 3: Mức thu nhập (Góc dưới trái: axes[1,0]) ---
# Quan trọng: Định nghĩa thứ tự hiển thị để cột không bị lộn xộn
income_order = ['Low', 'Medium', 'High']

# sns.countplot: Đếm số lượng và vẽ cột
# palette='viridis': Dùng bảng màu chuyển sắc (xanh -> vàng) cho đẹp
sns.countplot(data=data, x='Income_Level', order=income_order, palette='viridis', ax=axes[1, 0])
axes[1, 0].set_title('Phân bố Mức thu nhập (Income Level)')

# --- BIỂU ĐỒ 4: Sở thích du lịch (Góc dưới phải: axes[1,1]) ---
# Sắp xếp theo thứ tự phổ biến giảm dần (để cột dài nhất nằm trên cùng)
travel_order = data['Preferred_Travel_Type'].value_counts().index

# Lưu ý: Dùng tham số y='...' để vẽ biểu đồ nằm ngang (Horizontal)
sns.countplot(data=data, y='Preferred_Travel_Type', order=travel_order, palette='magma', ax=axes[1, 1])
axes[1, 1].set_title('Sở thích Du lịch (Preferred Travel Type)')

# Căn chỉnh tự động để các chữ không bị đè lên nhau
plt.tight_layout()
plt.show() # Hiển thị hình 1
```

**3.4 Phân tích mối tương quan tổng thể**


Để đánh giá mức độ liên kết tuyến tính giữa các đặc trưng quan trọng, ta thực hiện phân tích ma trận tương quan (Correlation Matrix) trên 5 biến chính, bao gồm: 3 biến định lượng (`Age`, `Spending`, `Duration`) và 2 biến phân loại sau khi mã hóa số (`Income`, `Repeat_Visitor`).

```python
# Bước 1: Tạo một bản sao dữ liệu để mã hóa tạm thời
df_corr = data.copy()

# Bước 2: Mã hóa các biến có thứ tự và nhị phân
# Income: Low=0, Medium=1, High=2
income_map = {'Low': 0, 'Medium': 1, 'High': 2}
df_corr['Income_Level'] = df_corr['Income_Level'].map(income_map)

# Repeat_Visitor: No=0, Yes=1
repeat_map = {'No': 0, 'Yes': 1}
df_corr['Repeat_Visitor'] = df_corr['Repeat_Visitor'].map(repeat_map)

# Bước 3: Chỉ chọn các cột số để tính tương quan
# (Lúc này Income và Repeat đã thành số nên sẽ được chọn)
cols_to_corr = ['Age', 'Spending_Amount', 'Avg_Trip_Duration', 'Income_Level', 'Repeat_Visitor']
correlation_matrix = df_corr[cols_to_corr].corr()

# Bước 4: Vẽ Heatmap
plt.figure(figsize=(10, 8))
plt.title('3.4. Ma trận tương quan (Correlation Matrix)', fontsize=16, fontweight='bold')

sns.heatmap(correlation_matrix, 
            annot=True,       # Hiển thị con số trong ô
            fmt=".2f",        # Lấy 2 số lẻ
            cmap='coolwarm',  # Màu Xanh (lạnh) - Đỏ (nóng)
            vmin=-1, vmax=1,  # Thang đo từ -1 đến 1
            linewidths=0.5, 
            square=True)

plt.tight_layout()
plt.show()
```

Việc các biến độc lập như thế này chứng tỏ mỗi biến đều mang lại một chiều không gian thông tin riêng biệt, giúp thuật toán phân cụm có cái nhìn đa chiều và toàn diện hơn về khách hàng.

- **Tiền xủ lý dữ liệu:** Các biến trong bộ dữ liệu độc lập và có đơn vị đo lường khác nhau (ví dụ: Tuổi tính theo năm, Chi tiêu theo nghìn). Nếu bỏ qua, các biến có giá trị lớn như Chi tiêu sẽ chi phối hoàn toàn khoảng cách Euclidean, khiến kết quả phân cụm bị sai lệch.

Do thuật toán phân cụm K-Means hoạt động dựa trên nguyên lý tính toán khoảng cách Euclidean giữa các điểm dữ liệu, độ chính xác của mô hình phụ thuộc hoàn toàn vào chất lượng và định dạng của dữ liệu đầu vào. Dựa trên đặc thù của 8 biến số trong bộ dữ liệu khách hàng, nhóm thực hiện quy trình tiền xử lý gồm hai giai đoạn chính: Mã hóa biến phân loại và Chuẩn hóa dữ liệu.

Trước khi tiến hành tiền xử lý, nhóm thực hiện tạo một bản sao của tập dữ liệu `data` và đặt tên là `data_processed`. Tất cả các bước xử lý dữ liệu đều được thực hiện trên bản sao này nhằm bảo đảm tính toàn vẹn của dữ liệu ban đầu, đồng thời cho phép đối chiếu và khôi phục khi cần thiết. Việc tạo bản sao cũng giúp tránh các sai lệch ngoài ý muốn do cơ chế tham chiếu trong Python, đồng thời tạo điều kiện thuận lợi cho việc thử nghiệm nhiều phương pháp tiền xử lý khác nhau trên cùng một nguồn dữ liệu. Đây là bước quan trọng nhằm đảm bảo tính tái lập và độ tin cậy cho toàn bộ quy trình phân tích.

```python
data_processed = data.copy()
data_processed.info()
```

**4.1. Mã hóa biến  loại  (Categorical Encoding)**

Dữ liệu thô chứa 5 biến định tính (categorical). Ta phân tích bản chất của từng biến để lựa chọn kỹ thuật mã hóa tối ưu nhất, cụ thể như sau:

**4.1.1. Xử lý biến có thứ tự (Ordinal Variable)**

**Biến:** `Income_Level` (Low, Medium, High)  
**Kỹ thuật:** *Ordinal Encoding*

**Quy trình:**  
Gán giá trị số theo thứ tự tăng dần:  
- Low = 0  
- Medium = 1  
- High = 2  

**Biện luận:**  
Biến thu nhập mang tính chất phân cấp rõ rệt. Việc sử dụng *Ordinal Encoding* giúp bảo toàn quan hệ thứ bậc tự nhiên (máy tính hiểu **High > Medium > Low**).  
Không sử dụng *One-Hot Encoding* vì kỹ thuật đó sẽ làm mất thông tin về thứ tự độ lớn giữa các mức.


```python
# Ordinal Encoding cho Income_Level
# Giữ thứ tự Low < Medium < High
income_order = {'Low': 0, 'Medium': 1, 'High': 2}
data_processed['Income_Level'] = data_processed['Income_Level'].map(income_order)
```

**4.1.2. Xử lý biến nhị phân (Binary Variable)**
- **Biến**: `Repeat_Visitor` (Yes, No)  
- **Kỹ thuật**: *Binary Mapping*

**Quy trình**:  
Chuyển đổi trực tiếp giá trị:
- `No = 0`  
- `Yes = 1`

**Biện luận**:  
Đối với biến chỉ có 2 giá trị loại trừ nhau, việc sử dụng One-Hot Encoding (tạo 2 cột) là không cần thiết và gây dư thừa dữ liệu (*redundancy*).  
Binary Mapping giúp:
- Tối ưu hóa không gian lưu trữ  
- Giảm số chiều dữ liệu (giữ nguyên 1 cột)  
- Vẫn đảm bảo tính toán khoảng cách chính xác trong các thuật toán phân cụm


```python
# Binary Mapping cho Repeat_Visitor
# Tối ưu 1 cột cho biến nhị phân
binary_map = {'No': 0, 'Yes': 1}
data_processed['Repeat_Visitor'] = data_processed['Repeat_Visitor'].map(binary_map)
```

**4.1.3. Xử lý biến định danh (Nominal Variables)**

- **Biến**: `Gender`, `Preferred_Travel_Type`, `Booking_Channel`  
- **Kỹ thuật**: *One-Hot Encoding*

**Quy trình**:  
Tách mỗi giá trị duy nhất của biến thành một cột nhị phân riêng biệt.  
Ví dụ: `Preferred_Travel_Type` được tách thành 4 cột:  
- `Adventure`  
- `Cultural`  
- `Leisure`  
- `Religious`

**Biện luận**:  
Các biến này là nhóm định danh ngang hàng, không có thứ tự hơn kém.  
Nếu gán số trực tiếp (1, 2, 3…), thuật toán có thể hiểu sai rằng các giá trị có mối quan hệ thứ bậc (như *Adventure < Cultural*).  
One-Hot Encoding loại bỏ vấn đề này bằng cách tạo ra **khoảng cách công bằng** giữa các nhóm và tránh đưa ý nghĩa thứ tự vào biến không có thứ tự.


```python
# One-Hot Encoding cho các biến định danh còn lại
# Các biến này không có thứ tự
nominal_vars = ['Gender', 'Preferred_Travel_Type', 'Booking_Channel']

data_processed_encoded = pd.get_dummies(
    data_processed,
    columns=nominal_vars,
    drop_first=False,
    dtype=int
)
```

Trước khi đến bước **CHUẨN HÓA (SCALING)** ta kiểm tra lại dữ liệu sau khi **MÃ HÓA (ENCODING)**: 

```python
print("Kích thước trước khi mã hóa (encoding): ",data.shape)
print("Kích thước sau khi mã hóa (encoding): ",data_processed_encoded.shape)
print("Tổng số giá trị thiếu sau khi mã hóa (encoding): ", data_processed_encoded.isnull().sum().sum() )
print("Thông tin các cột sau khi mã hóa (encoding) :")
print(data_processed_encoded.nunique())
display(data_processed_encoded.head())

```

Từ kết quả cho thấy toàn bộ dữ liệu đã được mã hóa thành dạng số hoàn chỉnh, không có giá trị thiếu và sẵn sàng cho bước **CHUẨN HÓA (Scaling)** nhằm đảm bảo các biến có thang đo khác nhau không gây sai lệch trong quá trình phân cụm 

**4.2. Chuẩn hóa dữ liệu (Feature Scaling)**

Sau bước mã hóa dữ liệu, các đặc trưng trong bộ dữ liệu có sự chênh lệch rất lớn về thang đo giá trị (Scale):

- **Biến `Spending_Amount`:** Có giá trị lớn nhất, dao động rộng từ *100.91 đến 4998.21*
- **Biến `Avg_Trip_Duration`:** Dao động từ *2 đến 20 ngày*
- **Biến `Age`**: Dao động từ *18 đến 69*  
- **Biến `Income_Level`**: Dao động từ *0 đến 2*  
- **Các biến One-Hot**: Chỉ nhận giá trị *0 hoặc 1*

**Vấn đề:** Thuật toán K-Means sử dụng khoảng cách Euclidean để xác định độ tương đồng. Nếu giữ nguyên thang đo gốc, biến **`Spending_Amount`** sẽ chi phối hoàn toàn việc tính toán khoảng cách, khiến thuật toán "coi thường" vai trò của các biến quan trọng khác.



 **Giải pháp**: Sử dụng *StandardScaler* (Z-score Normalization)

**Quy trình**:  
Áp dụng chuẩn hóa cho tập dữ liệu ( `Age`, `Spending_Amount`,`Avg_Trip_Duration` và `Income_Level` sau mã hóa):

$$z = \frac{x - \mu}{\sigma}$$



**Kết quả**

Tất cả các đặc trưng được đưa về phân phối chuẩn với:

- Trung bình: $\mu \approx 0$ 
- Độ lệch chuẩn: $\sigma = 1$

Điều này đảm bảo **tính công bằng về trọng số đóng góp** của tất cả các biến (Tuổi, Thu nhập, Loại hình du lịch, Booking Channel,…) trong việc hình thành các cụm khách hàng.


```python
# CHUẨN HÓA (SCALING)
# Lấy danh sách các cột
features = ['Age','Spending_Amount','Avg_Trip_Duration','Income_Level']
#Tạo bản sao
data_processed_scaled = data_processed_encoded.copy()
# Tính Mean và Std cho các cột
means = data_processed_scaled[features].mean()
stds = data_processed_scaled[features].std()
# Áp dụng công thức:
data_processed_scaled[features] = (data_processed_scaled[features] - means)/stds

display(data_processed_scaled.head())
```

Quan sát bảng dữ liệu sau xử lý, ta thấy chiến lược chuẩn hóa có chọn lọc đã được thực hiện hiệu quả: các biến định lượng là `Age`,`Spending_Amount`,`Avg_Trip_Duration` và `Income_Level` đã được đưa về phân phối chuẩn với các giá trị dao động quanh 0, giúp loại bỏ sự chênh lệch về độ lớn đơn vị; trong khi đó, các biến định danh và nhị phân (như Gender, Repeat_Visitor) vẫn giữ nguyên giá trị 0 và 1. Cấu trúc dữ liệu này đảm bảo tính công bằng về trọng số đóng góp của tất cả các đặc trưng khi tính toán khoảng cách Euclidean, sẵn sàng cho việc đưa vào mô hình phân cụm K-Means

**4.3. Tổng kết Tiền xử lý dữ liệu**

Để đảm bảo tính chính xác cho thuật toán K-Means (vốn dựa trên khoảng cách Euclidean), nhóm đã phân tích kỹ lưỡng đặc tính của từng biến số và áp dụng chiến lược tiền xử lý tối ưu nhất. Dưới đây là bảng tổng hợp các kỹ thuật đã cân nhắc và quyết định lựa chọn.





**Bảng Tổng hợp kỹ thuật tiền xử lý và biện luận lựa chọn**

| Biến số (Variable) | Loại dữ liệu | Các kỹ thuật khả thi | Kỹ thuật được chọn | Lý do tối ưu (Biện luận) |
| :--- | :--- | :--- | :--- | :--- |
| **Age**<br>**Spending_Amount**<br>**Avg_Trip_Duration** | Định lượng<br>(Liên tục) |- Min-Max Scaling<br>- StandardScaler <br> | **StandardScaler** | Các biến có thang đo lớn hơn nhiều so với các biến đã mã hóa (0-1). StandardScaler đưa biến về phân phối chuẩn ($\mu=0, \sigma=1$), giúp loại bỏ sự chênh lệch đơn vị đo lường và hạn chế ảnh hưởng của giá trị ngoại lai tốt hơn Min-Max Scaling |
| **Income_Level** | Định tính<br>(Có thứ tự) | - One-Hot Encoding<br>- Ordinal Encoding | **Ordinal Encoding + StandardScaler** | Dữ liệu có tính phân cấp (Low < Medium < High). Ordinal Encoding bảo toàn thứ tự này. Sau đó áp dụng StandardScaler để đồng bộ trọng số với biến `Age`,`Spending_Amount` và `Avg_Trip_Duration`. |
| **Repeat_Visitor** | Nhị phân<br>(Yes/No) | - One-Hot Encoding<br>- Binary Mapping | **Binary Mapping** | Vì biến chỉ có 2 giá trị loại trừ nhau, One-Hot Encoding (tạo 2 cột) gây dư thừa dữ liệu. Binary Mapping (giữ 1 cột 0/1) tối ưu hóa không gian lưu trữ mà vẫn đủ thông tin. |
| **Gender**<br>**Travel_Type**<br>**Booking_Channel** | Định danh<br>(Không thứ tự) | - Label Encoding<br>- One-Hot Encoding | **One-Hot Encoding** | Các giá trị trong nhóm này hoàn toàn bình đẳng, không có thứ tự hơn kém (Ví dụ: Adventure không lớn hơn Cultural). Label Encoding sẽ khiến thuật toán hiểu sai về thứ bậc. One-Hot Encoding tạo ra các vector trực giao, đảm bảo khoảng cách công bằng giữa các nhóm. |

```python
X = data_processed_scaled 
print("\n--- Dữ liệu hoàn chỉnh (X) sẵn sàng cho K-Means ---")
print(f"Kích thước cuối cùng: {X.shape}")
display(X.describe().round(2))
```

**CHƯƠNG 5: LỰA CHỌN THUẬT TOÁN VÀ HUẤN LUYỆN MÔ HÌNH**

Sau khi hoàn tất quá trình tiền xử lý, dữ liệu được đưa vào giai đoạn xây dựng mô hình phân cụm. Chương này trình bày cơ sở lựa chọn thuật toán, quy trình xác định số cụm tối ưu và kết quả huấn luyện mô hình.

**5.1. Lựa chọn thuật toán phân cụm**

Trong Machine Learning, có hàng chục thuật toán phân cụm, nhưng chúng thường được chia thành 4 nhóm chính:

| Thuật toán | Ưu điểm (Pros) | Nhược điểm (Cons) | Phù hợp nhất khi nào? |
|-----------|----------------|-------------------|------------------------|
| **K-Means** | - Tốc độ nhanh, tính toán đơn giản. <br> - Dễ giải thích ý nghĩa kinh doanh (Marketing). <br> - Hiệu quả với tập dữ liệu lớn. | - Phải chọn số cụm K trước. <br> - Nhạy cảm với nhiễu (outliers). <br> - Chỉ tìm được cụm dạng hình cầu. | Phân khúc khách hàng, nén ảnh, phân loại văn bản đơn giản. |
| **Hierarchical Clustering** | - Không cần chọn số cụm trước. <br> - Trực quan hóa cấu trúc dữ liệu tốt (dendrogram). | - Rất chậm với dữ liệu lớn (O(N³)). <br> - Sau khi gom cụm thì không tách được nữa. | Tập dữ liệu nhỏ, cần phân tích cấu trúc phân cấp (ví dụ: phả hệ sinh học). |
| **DBSCAN** | - Không cần chọn số cụm trước. <br> - Xử lý tốt cụm có hình dạng phức tạp. <br> - Tự động loại bỏ nhiễu. | - Khó chọn tham số khoảng cách <br> - Hoạt động kém nếu mật độ cụm không đều. | Dữ liệu địa lý, nhận dạng hình ảnh, dữ liệu có nhiều nhiễu. |
| **GMM (Gaussian Mixture Model)** | - Cụm linh hoạt (hình elip). <br> - Cho biết xác suất thuộc về cụm (soft clustering). | - Tính toán phức tạp, chậm. <br> - Dễ rơi vào  điểm tối ưu cục bộ. | Khi cần biết độ chắc chắn của phân loại (ví dụ: phát hiện gian lận tài chính). |


Mặc dù có nhiều thuật toán phân cụm tiên tiến như **DBSCAN** hay **Hierarchical Clustering**, nhóm quyết định lựa chọn **K-Means** vì các lý do sau:

1. **Mục tiêu kinh doanh:**
Trong bài toán **phân khúc khách hàng (Customer Clustering)**, mục tiêu là phân loại **toàn bộ khách hàng** vào các nhóm để đề xuất chiến lược Marketing.  
K-Means đảm bảo **mọi điểm dữ liệu đều được phân cụm**, trong khi DBSCAN có thể loại bỏ một số khách hàng “khác biệt” và coi là *nhiễu (Noise)* — điều này không phù hợp cho việc chăm sóc khách hàng.

2. **Khả năng giải thích (Interpretability):**
K-Means tạo ra các **tâm cụm (Centroids)** đại diện rõ ràng, giúp dễ dàng mô tả chân dung khách hàng (ví dụ: nhóm có thu nhập cao, độ tuổi thấp).  
Điều này thân thiện hơn với bộ phận kinh doanh so với độ phức tạp của **GMM** hay **DBSCAN**.

3. **Hiệu suất:**
Với bộ dữ liệu đã được **One-Hot Encoding** (số chiều tăng lên), K-Means hoạt động **ổn định và nhanh** hơn đáng kể so với **Hierarchical Clustering**, vốn tiêu tốn rất nhiều tài nguyên tính toán và không phù hợp cho dữ liệu tầm trung hoặc lớn.

4. **Tính ổn định:**
Với dữ liệu đã được chuẩn hóa (StandardScaler) và mã hóa kỹ lưỡng ở Chương 4, **K-Means** hoạt động rất hiệu quả trong việc tìm ra các nhóm có đặc điểm tương đồng.



**5.2. Xác định số lượng cụm tối ưu (K)**

**5.2.1. Phương pháp thực hiện**

Dựa trên cơ sở lý thuyết đã trình bày tại **Chương 2 (Mục 2.4.2)**, nhóm tiến hành thực nghiệm trên tập dữ liệu đã tiền xử lý ($X$) để tìm ra số cụm $K$ phù hợp nhất. Quy trình thực hiện gồm các bước sau:

1. **Thiết lập phạm vi thử nghiệm:** 
Chạy thuật toán **K-Means** với số cụm $K$ thay đổi từ **2 đến 10**.

2. **Ghi nhận chỉ số:**
Tại mỗi giá trị $K$, tính toán và lưu trữ hai chỉ số đánh giá:
- Inertia (WCSS): Dùng để vẽ biểu đồ *Elbow*, đánh giá độ nén của cụm.
- Silhouette Score: Đánh giá mức độ tách biệt trung bình giữa các cụm.
3. **Trực quan hóa:**
Biểu diễn đồng thời **hai chỉ số** trên cùng một biểu đồ *(Dual-Axis Chart)* nhằm quan sát sự cân bằng tối ưu giữa:
- Độ nén của cụm (WCSS)
- Độ tách biệt của các cụm (Silhouette Score)



```python
inertia = []
silhouette_scores = []
K_range = range(2, 11)  # Thử số cụm từ 2 đến 10

for k in K_range:
    # Khởi tạo và chạy KMeans
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    
    # Lưu giá trị Inertia (cho Elbow)
    inertia.append(kmeans.inertia_)
    
    # Lưu giá trị Silhouette Score
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)

# --- VẼ BIỂU ĐỒ KẾT HỢP ---
fig, ax1 = plt.subplots(figsize=(12, 6))

# Trục 1 (Trái): Elbow - Inertia
color = 'tab:red'
ax1.set_xlabel('Số lượng cụm (K)', fontsize=12)
ax1.set_ylabel('Inertia (Elbow)', color=color, fontsize=12)
ax1.plot(K_range, inertia, color=color, marker='o', linewidth=2, label='Inertia')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, linestyle='--', alpha=0.6)

# Trục 2 (Phải): Silhouette Score
ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('Silhouette Score', color=color, fontsize=12)
ax2.plot(K_range, silhouette_scores, color=color, marker='s', linestyle='--', linewidth=2, label='Silhouette')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Phương pháp kết hợp Elbow và Silhouette để chọn K', fontsize=14)
plt.show()

```

**5.2.2. Kết quả và Biện luận chọn K**

Kết quả thực nghiệm được thể hiện tại **Biểu đồ** trên. Dựa trên phân tích kỹ lưỡng hai chỉ số, nhóm quyết định lựa chọn số cụm tối ưu là $K=5$.

**Cơ sở biện luận:**
1. **Theo chỉ số Silhouette (Đường nét đứt màu xanh):**
    - Chỉ số Silhouette đo lường mức độ tách biệt giữa các cụm (càng cao càng tốt).
    - **Tại K=2:** Chỉ số cao nhất (~0.113). Về mặt toán học thuần túy, đây là cách chia tốt nhất. Tuy nhiên, trong bài toán phân khúc khách hàng du lịch, việc chia thành 2 nhóm thường là quá ít và quá rộng, không đủ chi tiết để làm Marketing hiệu quả.
    - **Tại K=5:** Ta có thể thấy một đỉnh cục bộ (local peak) rõ rệt. Sau khi giảm sâu ở K=3 và tăng nhẹ ở K=4, chỉ số này vọt lên cao nhất tại K=5 (xấp xỉ 0.107) trước khi giảm xuống ở K=6. Đây là tín hiệu mạnh cho thấy việc chia 5 cụm tạo ra cấu trúc dữ liệu tốt hơn so với 3, 4 hoặc 6 cụm.
2. **Theo phương pháp Elbow (Đường màu đỏ):**
    - Chỉ số này đo độ nén của cụm (càng thấp càng tốt, nhưng cần tìm điểm gãy "khủy tay").
    - Đường màu đỏ giảm khá đều và mượt, không có một "khủy tay" vuông góc rõ rệt (điều này thường xảy ra với dữ liệu hỗn hợp gồm cả số và phân loại).
    - Tuy nhiên, tại khoảng K=4 đến K=6, tốc độ giảm của quán tính bắt đầu chậm lại đôi chút so với đoạn đầu. Kết hợp với tín hiệu từ Silhouette, K=5 nằm ở vùng chấp nhận được về độ nén dữ liệu.
3. **Góc độ kinh doanh:**
    - Với bộ dữ liệu du lịch, 5 nhóm khách hàng (Personas) thường là con số lý tưởng để xây dựng các chiến lược bán hàng.
    - Nếu chọn K=2, phân khúc quá sơ sài.
    - Nếu chọn K=7 (một đỉnh nhỏ khác của Silhouette), việc quản lý 7 chiến lược marketing có thể trở nên phức tạp và các nhóm có thể bị trùng lặp nhiều.
    
**Kết luận:** Nhóm quyết định thiết lập số cụm $K=5$ cho mô hình phân khúc khách hàng.

**5.3. Huấn luyện mô hình**

Sau khi xác định được số cụm tối ưu là $K=5$, nhóm tiến hành thiết lập và huấn luyện mô hình K-Means chính thức trên tập dữ liệu đã chuẩn hóa.

```python
# --- HUẤN LUYỆN MÔ HÌNH FINAL ---

optimal_k = 5  

print(f"Tiến hành huấn luyện mô hình với K = {optimal_k}...")

# Khởi tạo mô hình chính thức
kmeans_final = KMeans(n_clusters=optimal_k, init='k-means++', n_init=10, random_state=42)

# Huấn luyện (Fit)
kmeans_final.fit(X)

# Lấy nhãn cụm (Labels)
cluster_labels = kmeans_final.labels_

# --- GÁN NHÃN VÀO DỮ LIỆU ---
# Quan trọng: Gán nhãn vào dữ liệu gốc (data) để phân tích, không gán vào X (vì X đã bị biến đổi khó đọc)
df_result = data.copy() # data là dữ liệu gốc ban đầu
df_result['Cluster_Label'] = cluster_labels

print("--- Huấn luyện hoàn tất! ---")
print("Mẫu dữ liệu sau khi phân cụm:")
display(df_result.head())

# Kiểm tra số lượng khách hàng trong từng cụm
print("\nPhân bố số lượng khách hàng trong các cụm:")
print(df_result['Cluster_Label'].value_counts().sort_index())
```

  - **Kênh:** Báo điện tử, Facebook (nhắm mục tiêu độ tuổi 50+), Zalo.


**Cụm 1: "Nhóm Trung Niên Hào Phóng"**  
- **Đặc điểm nhận dạng:**
  - Tuổi: Trung niên (~56 tuổi).
  - Thu nhập: Trung bình (Medium) - Nhưng điểm thú vị là họ chi tiêu mạnh tay nhất.
  - Chi tiêu: Cao (~3.887 - Cao nhất bảng).
  - Hành vi: Thích văn hóa, đặt qua Đại lý (Agent).
- **Insight:** Dữ liệu cho thấy một sự "nghịch lý" thú vị khi nhóm này có thu nhập trung bình nhưng lại có mức chi tiêu cao nhất trong tất cả các nhóm. Điều này chỉ ra rằng họ coi du lịch là một khoản đầu tư lớn cho đời sống tinh thần (Văn hóa). Việc họ phụ thuộc hoàn toàn vào Đại lý (Agent) chứng tỏ đây là nhóm khách hàng mua sự tiện lợi và dịch vụ, họ sẵn sàng trả phí cao để được phục vụ trọn gói và tránh các rủi ro hoặc phiền phức trong khâu tổ chức.
- **Chiến lược Marketing:**
  - **Thông điệp:** "Dịch vụ đẳng cấp - Trải nghiệm văn hóa chuyên sâu".
  - **Sản phẩm:** Tour thiết kế riêng, hướng dẫn viên am hiểu lịch sử, khách sạn 4-5 sao mang đậm bản sắc địa phương.
  - **Kênh:** Hợp tác chặt chẽ với các Đại lý du lịch (Travel Agency), Telesales chăm sóc khách hàng VIP.


**Cụm 2: "Nhóm Tuổi Trẻ 'Chịu Chơi'"**  
- **Đặc điểm nhận dạng:**
  - Tuổi: Trẻ (~31 tuổi).
  - Thu nhập: Trung bình (Medium).
  - Chi tiêu: Khá cao (~3.020 - Đứng thứ 2).
  - Hành vi: Mạo hiểm, đi ngắn ngày nhất (~6 ngày).
- **Insight:** Đặc trưng cốt lõi của nhóm này là "Cường độ chi tiêu cao trong thời gian ngắn". Với thời gian chuyến đi ngắn nhất (~6 ngày) nhưng mức chi tiêu lại thuộc top đầu, có thể thấy đây là nhóm khách hàng bận rộn ("time-poor") nhưng có khả năng tài chính. Họ chấp nhận chi trả mạnh tay để tối đa hóa trải nghiệm cảm giác mạnh (Adventure) trong khoảng thời gian nghỉ phép ít ỏi. Động lực tiêu dùng của họ là sự bứt phá và kịch tính thay vì thư giãn đơn thuần.
- **Chiến lược Marketing:**
  - **Thông điệp:** "Chinh phục thử thách - Bứt phá giới hạn".
  - **Sản phẩm:** Tour trekking cao cấp, nhảy dù, lặn biển, thám hiểm hang động (được tổ chức chuyên nghiệp, an toàn cao).
  - **Kênh:** Video ngắn (TikTok, Reels, Youtube Shorts) thể hiện sự kịch tính.


**Cụm 3: "Nhóm Trung Niên Tối Giản"**  
- **Đặc điểm nhận dạng:**
  - Tuổi: Trung niên (~42 tuổi).
  - Thu nhập: Trung bình (Medium).
  - Chi tiêu: Thấp nhất (~1.300).
  - Hành vi: Du lịch Tôn giáo.
- **Insight:** Đây là nhóm có hành vi tiêu dùng tách biệt hoàn toàn với khả năng tài chính. Dù có thu nhập ổn định (Medium), mức chi tiêu của họ lại chạm đáy. Nguyên nhân nằm ở loại hình du lịch Tôn giáo, vốn đề cao sự thanh tịnh và giản đơn. Hành vi này cho thấy nhu cầu của họ tập trung vào giá trị tinh thần và kết nối cộng đồng hơn là thụ hưởng cơ sở vật chất. Họ là nhóm ít nhạy cảm với các dịch vụ nâng cao nhưng lại rất trung thành với mục đích chuyến đi.
- **Chiến lược Marketing:**
  - **Thông điệp:** "Hành trình tâm linh - Về miền thanh tịnh".
  - **Sản phẩm:** Tour hành hương giá rẻ, dịch vụ xe đưa đón tận nơi đến các điểm tâm linh. Kết nối cộng đồng.
  - **Kênh:** Các hội nhóm cộng đồng, Word-of-mouth (Truyền miệng) rất hiệu quả với nhóm này.


**Cụm 4: "Nhóm Giới Trẻ Ưu Tiên Trải Nghiệm"**  
- **Đặc điểm nhận dạng:**
  - Tuổi: Trẻ (~32 tuổi).
  - Thu nhập: Thấp (Low).
  - Chi tiêu: Khá cao (~2.866).
  - Hành vi: Dùng App, đi dài ngày nhất (~15 ngày).
- **Insight:** Nhóm này thể hiện rõ xu hướng tiêu dùng của thế hệ Gen Y/Z: "Trải nghiệm là ưu tiên số 1". Mặc dù bị hạn chế về thu nhập (Low Income), họ vẫn dành một tỷ trọng lớn ngân sách và thời gian cho du lịch. Việc là nhóm duy nhất sử dụng App cho thấy họ là những người "nghiện" công nghệ, có khả năng săn tìm các ưu đãi thông minh để kéo dài chuyến đi. Với họ, du lịch là một phần của phong cách sống (Lifestyle) chứ không phải là xa xỉ phẩm.
- **Chiến lược Marketing:**
  - **Thông điệp:** "Sống trọn đam mê - Vi vu không nghĩ ngợi".
  - **Sản phẩm:** Các gói "Staycation", combo vé máy bay + khách sạn linh hoạt, tích hợp nhiều voucher ăn uống/vui chơi trên App.
  - **Kênh:** Mobile App (Thông báo đẩy, Gamification), Influencer Marketing (KOLs du lịch).

**CHƯƠNG 7: KẾT LUẬN VÀ KIẾN NGHỊ**

**7.1. Tổng kết kết quả nghiên cứu (Conclusion)**

Đồ án đã thực hiện trọn vẹn quy trình khai phá dữ liệu khách hàng du lịch, từ bước tiền xử lý, phân tích khám phá (EDA) đến việc xây dựng mô hình phân cụm K-Means. Các kết quả chính đạt được bao gồm:

- **Xây dựng thành công quy trình phân cụm :** Đồ án xây dựng thành công quy trình phân cụm dữ liệu khách hàng du lịch từ bước xử lý làm sạch dữ liệu, phân tích khám phá (EDA) đến việc huấn luyện mô hình học máy K-Means với số cụm tối ưu K=5.

- **Định danh chính xác 5 phân khúc khách hàng đặc trưng:** Từ tập dữ liệu hỗn loạn ban đầu, đồ án đã tách biệt và phác họa rõ nét 5 chân dung khách hàng mục tiêu:
  - Cụm 0: Nhóm Hưu Trí Bình Dân
  - Cụm 1: Nhóm Trung Niên Hào Phóng
  - Cụm 2: Nhóm Tuổi Trẻ 'Chịu Chơi'
  - Cụm 3: Nhóm Trung Niên Tối Giản
  - Cụm 4: Nhóm Giới Trẻ Ưu Tiên Trải Nghiệm

- **Tri thức (Insight):** Đồ án đã bác bỏ giả định rằng "Thu nhập quyết định tất cả". Thay vào đó, Độ tuổi và Hành vi chi tiêu mới là hai yếu tố then chốt phân định khách hàng. Năm chân dung khách hàng được định danh (từ Hưu Trí Bình Dân đến Giới Trẻ Ưu Tiên Trải Nghiệm) là cơ sở vững chắc để doanh nghiệp chuyển đổi từ tiếp thị đại trà sang tiếp thị cá nhân hóa.

- **Đề xuất giải pháp thực tiễn:** Dựa trên đặc điểm của từng cụm, đồ án đã đưa ra các chiến lược kinh doanh cụ thể theo mô hình Marketing Mix, giúp doanh nghiệp thay đổi tư duy từ tiếp thị đại trà (Mass Marketing) sang tiếp thị cá nhân hóa (Personalized Marketing), tối ưu hóa nguồn lực và gia tăng trải nghiệm khách hàng.

**7.2 Đóng góp của đồ án**
- **Về mặt lý thuyết:** Hệ thống hóa quy trình ứng dụng Machine Learning trong ngành du lịch, cung cấp một khung tham chiếu (framework) cho các nghiên cứu tương tự.
- **Về mặt thực tiễn:** Cung cấp cho doanh nghiệp một "tấm bản đồ" chi tiết về thị trường khách hàng hiện tại. Các chiến lược đề xuất cho từng cụm (như tập trung vào nhóm VIP Cụm 1 hay khai thác nhóm Trẻ Cụm 4 qua App) có tính khả thi cao và tác động trực tiếp đến doanh thu.

**7.3 Hạn chế của đồ án (Limitations)**

Mặc dù đã đạt được những kết quả khả quan, đồ án vẫn tồn tại một số hạn chế nhất định cần được nhìn nhận một cách khách quan:

- **Về khía cạnh Dữ liệu:**
  - Kích thước dữ liệu dừng lại ở mức trung bình (~2.500 dòng), có thể chưa đại diện đầy đủ cho toàn bộ tính chất phức tạp của thị trường du lịch thực tế.
  - Dữ liệu mang tính chất tĩnh (Snapshot), thiếu các yếu tố biến động theo thời gian như tính mùa vụ (mùa cao điểm/thấp điểm) hoặc các dữ liệu hành vi thời gian thực (real-time).

- **Về khía cạnh Thuật toán:**
  - Thuật toán K-Means giả định các cụm có hình cầu (spherical) và kích thước tương đồng. Điều này có thể chưa tối ưu nếu các nhóm khách hàng trong thực tế có hình dạng phân bố phức tạp hoặc mật độ chênh lệch quá lớn.
  - Việc lựa chọn số cụm K phụ thuộc nhiều vào các phương pháp heuristic (như Elbow Method), mang tính tương đối.

- **Về tính Thực tiễn:**
  - Các đề xuất chiến lược kinh doanh mới chỉ dừng lại ở mức giả thuyết dựa trên dữ liệu quá khứ. Hiệu quả thực tế chưa được kiểm chứng qua các phương pháp thực nghiệm như A/B Testing hoặc đo lường doanh thu sau triển khai.

**7.4 Hướng phát triển (Future Work)**

Để khắc phục các hạn chế trên và nâng cao giá trị ứng dụng của đồ án, các hướng phát triển tiếp theo được đề xuất bao gồm:

- **Mở rộng phạm vi dữ liệu:** Tích hợp thêm dữ liệu hành vi người dùng trên nền tảng số (Digital Footprints) như: lịch sử click chuột, thời gian xem trang, từ khóa tìm kiếm và các dữ liệu phản hồi phi cấu trúc (Review text) để thấu hiểu sâu hơn tâm lý khách hàng.

- **Nâng cấp mô hình thuật toán:** Thử nghiệm và so sánh hiệu quả với các thuật toán phân cụm nâng cao hơn như DBSCAN (để xử lý tốt hơn các dữ liệu nhiễu và cụm có hình dạng bất kỳ) hoặc Hierarchical Clustering (để xây dựng phân cấp khách hàng chi tiết).

- **Xây dựng Hệ thống Gợi ý (Recommendation System):** Phát triển ứng dụng thực tế bằng cách tích hợp kết quả phân cụm vào hệ thống gợi ý tự động. Ví dụ: Hệ thống sẽ tự động hiển thị tour "Mạo hiểm" ngay khi nhận diện khách hàng truy cập có đặc điểm thuộc Cụm 2, giúp cá nhân hóa trải nghiệm người dùng ngay từ điểm chạm đầu tiên.
